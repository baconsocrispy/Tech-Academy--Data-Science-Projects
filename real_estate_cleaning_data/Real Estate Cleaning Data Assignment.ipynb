```python
(data.houseAge - data.houseAge.mean()) / data.houseAge.std() # standardize houseAge column data
```

```python
standard = (data.houseAge - data.houseAge.mean()) / data.houseAge.std() # set standardized series to variable
```

```python
standard # confirm data
```

```python
normalize = (standard - standard.min()) / (standard.max() - standard.min()) # normalize series and set to variable
```

```python
normalize # confirm result
```

```python
normalize.min() # check min == 0
```

```python
normalize.max() # check max == 1
```

```python
data['houseAgeNormalized'] = normalize # create a new column with normalized data
```

```python
data # confirm data
```

```python
data.drop('numberOfConvenienceStores', axis=1, inplace=True) # remove column per assignment
```
